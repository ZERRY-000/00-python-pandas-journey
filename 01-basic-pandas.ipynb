# Basic Aggregation 
## Data Setup

```python
import pandas as pd

data = {
    "name":       ["Alice", "Bob", "Charlie", "Diana", "Eve", "Frank"],
    "department": ["HR", "IT", "IT", "Finance", "HR", "Finance"],
    "salary":     [55000, 72000, 68000, 80000, 52000, 91000],
    "age":        [29, 34, 28, 41, 35, 47]
}

df = pd.DataFrame(data)
```

## Basic Aggregation (whole column)

```python
print("Salary summary:", df["salary"].sum())    # total
print("Salary mean:",df["salary"].mean())   # average
print("Salary minimum:",df["salary"].min())    # minimum
print("Salary maximum:",df["salary"].max())    # maximum
print("Salary non-null row count:",df["salary"].count())  # count non-null
print("Salary standard deviation (S.D.):",df["salary"].std())    # standard deviation
```

## GROUP BY + single aggregate and multiple aggregates

```python
print(df.groupby("department")["salary"].mean())
print(df.groupby("department")["salary"].sum())
print(df.groupby("department")["salary"].count())

print(
df.groupby("department").agg(
    avg_salary = ("salary", "mean"),
    max_salary = ("salary", "max"),
    headcount  = ("name",   "count")
)
)

```

## GROUP BY multiple columns


```python
df.groupby(["department", "age"])["salary"].mean()
```

## HAVING (filter after group by)

```python
result = df.groupby("department")["salary"].mean()
result[result > 65000]   # only departments with avg salary > 65000
```

## value_counts (like COUNT + GROUP BY in one shot)

```python
df["department"].value_counts()
```

## describe() — all stats at once

```python
df.groupby("department")["salary"].describe()
```
